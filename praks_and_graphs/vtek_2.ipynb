```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

Секундомер

```python
x_train = torch.linspace(0, 9, 10)
y_train = torch.tensor([9.2, 9.636666667, 10.47666667, 10.50666667, 11.13333333, 11.53, 11.92333333, 12.29333333, 12.78, 13.02])/10
y_train = torch.pow(y_train, 2)
errors = [0.021863, 0.006005, 0.006585, 0.0067283, 0.011925, 0.008024, 0.012258, 0.003132, 0.021999, 0.007679]
```

```python
print(y_train.shape)
print(x_train.shape)
```

```python
plt.scatter(x_train, y_train, label = 'train', )
#plt.scatter(x_train, y_pred.detach(), label = 'prediction')
plt.legend(loc = 'upper left')
plt.title('training and predicted data')
plt.grid(True)
plt.show()
```

```python
plt.errorbar(x_train, y_train, errors, fmt = 'o')
line = y_train[0]+ (y_train[-1]-y_train[0])/(x_train[-1]-x_train[0])*x_train
plt.plot(x_train, line)
#plt.scatter(x_train, y_pred.detach(), label = 'prediction')
plt.legend(loc = 'upper left')
plt.title('Зависимость T\u00B2(m) для секундомера')
plt.ylabel('T\u00B2(m), с\u00B2')
plt.xlabel('m, шт')
plt.grid(True)
plt.savefig('vtek_2_1_1')
plt.show()
print("A = ",(y_train[-1]-y_train[0])/(x_train[-1]-x_train[0]))
```
